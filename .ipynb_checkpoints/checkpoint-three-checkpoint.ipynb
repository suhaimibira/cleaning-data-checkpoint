Import the necessary libraries and create your dataframe(s).

```python
import pandas as pd 
import numpy as np 

features_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/Features data set.csv')
sales_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/sales data-set.csv')
stores_df = pd.read_csv('C:/Users/Sabria/Desktop/Python/Database for GA/stores data-set.csv')

combined_df = pd.concat([features_df, sales_df, stores_df], axis=1)
combined_df.reset_index(drop=True, inplace=True)
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
print(combined_df.head(25).isna())
print(combined_df.isna().sum())
print(combined_df.isnull().sum())

print(combined_df.shape)
combined_df_cleaned= combined_df.dropna()
print(combined_df_cleaned.shape)
# dropping rows with Naan
```

## Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
numeric_df = combined_df.select_dtypes(include=[np.number])

mean = np.mean(numeric_df, axis=0)
std_dev = np.std(numeric_df, axis=0)

z_scores = (numeric_df - mean)/std_dev
outliers = np.abs(z_scores) > 3

print("Z-scores:\n", z_scores)
print("\nOutliers:\n", outliers)
```

## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

```python
missing_percentages = combined_df.isna().mean()
print(missing_percentages)
# there is over 98% for the majority of the data on the top. 

duplicate_rows = combined_df[combined_df.duplicated()]
print("Duplicate rows:\n", duplicate_rows)
```

## Inconsistent Data

Check for inconsistent data and address any that arises. As always, use code comments to illustrate your thought process.

```python
#basically making sure the column is numeric and if a non-numeric value is found = it converts into Naan
inconsistent_data_fuel_price = pd.to_numeric(combined_df['Fuel_Price'], errors='coerce').isna()
inconsistent_data_temperature= pd.to_numeric(combined_df['Temperature'], errors='coerce').isna()
inconsistent_data_CPI = pd.to_numeric(combined_df['CPI'], errors='coerce').isna()
inconsistent_data_unemployment = pd.to_numeric(combined_df['Unemployment'], errors='coerce').isna()
inconsistent_data_size = pd.to_numeric(combined_df['Size'], errors='coerce').isna()
inconsistent_data_weekly_sales= pd.to_numeric(combined_df['Weekly_Sales'], errors='coerce').isna()

print("Inconsistent data in 'Fuel Price':\n", combined_df[inconsistent_data_fuel_price])
print("Inconsistent data in 'Temperature':\n", combined_df[inconsistent_data_temperature])
print("Inconsistent data in 'CPI':\n", combined_df[inconsistent_data_CPI])
print("Inconsistent data in 'Unemployment':\n", combined_df[inconsistent_data_unemployment])
print("Inconsistent data in 'Size':\n", combined_df[inconsistent_data_size])
print("Inconsistent data in 'Weekly Sales':\n", combined_df[inconsistent_data_weekly_sales])

#checking to see if there are duplicated columns and removing them except the one instance. 
print(combined_df.columns[combined_df.columns.duplicated()])
cleaned_combined_df = combined_df.loc[:, ~combined_df.columns.duplicated()]
print(cleaned_combined_df)

#fixing the 'Date' column and replacing empty spaces with 'NaT'
combined_df['Date_Fixed']= pd.to_datetime(combined_df['Date'], errors='coerce')
```
